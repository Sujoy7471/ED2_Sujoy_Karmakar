## Question 2:

#### Part a:

Here in x_ct convention, to draw the x' and ct' axes, we use the equations as,  
 * For x' axis, ct' = 0. So, using the transformation equations, (ct) = $\frac{v}{c}x$
 * For ct' axis, x' = 0. So, using the transformation equations, (ct) = $\frac{c}{v}x$

Here in x_ict convention, to draw the x' and ict' axes, we use the equations as,  
 * For x' axis, ict' = 0. So, using the transformation equations, (ict) = $i\frac{v}{c}x$
 * For ict' axis, x' = 0. So, using the transformation equations, (ict) = $-i\frac{c}{v}x$

```python
## Defining the plot function

def plot_axes(beta):
    # x-ct convention
    x = np.linspace(-3, 3, 500)
    ct = np.linspace(-3, 3, 500)
    ctp_axis = beta * ct       # ct'-axis
    xp_axis = beta * x        # x'-axis

    # x-ict convention
    ict = np.linspace(-3, 3, 500)
    ctp_axis_ict = beta * ict  # ict'-axis
    ict_xp_axis = -beta * x      # x'-axis

    # make 2 plots
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))

    # plot for x-ct
    axes[0].plot(ctp_axis, ct, label="ct'-axis")
    axes[0].plot(x, xp_axis, label="x'-axis")
    axes[0].axhline(0, color='k', linewidth=0.5)
    axes[0].axvline(0, color='k', linewidth=0.5)
    axes[0].set_xlim(-3, 3)
    axes[0].set_ylim(-3, 3)
    axes[0].set_xlabel("x")
    axes[0].set_ylabel("ct")
    axes[0].legend()
    axes[0].set_title(rf"x-ct convention for $beta$ = {beta}" )

    # plot for x-ict
    axes[1].plot(ctp_axis_ict, ict, label="ct'-axis")
    axes[1].plot(x, ict_xp_axis, label="x'-axis")
    axes[1].axhline(0, color='k', linewidth=0.5)
    axes[1].axvline(0, color='k', linewidth=0.5)
    axes[1].set_xlim(-3, 3)
    axes[1].set_ylim(-3, 3)
    axes[1].set_xlabel("x")
    axes[1].set_ylabel("ict")
    axes[1].legend()
    axes[1].set_title(rf"x-ict convention for $beta$ = {beta}")

    plt.tight_layout()
    plt.show()

## Plotting the plots
beta = [0.01, 0.5, 0.9999]

for i in range(len(beta)):
    plot_axes(beta[i])
```

```python
vs = np.arange(0, 1, 0.001)
alpha = np.arange(0, np.pi, 0.5)

## Plot
plt.figure(figsize=(8, 8))
for i in range(len(alpha)):
    plt.plot(vs, (vs*np.sin(alpha[i]) / (1 - vs*np.cos(alpha[i]))), label=rf"$\alpha$ = {alpha[i]:.2f}")

plt.xlabel(r"$\frac{V_s}{c}$", fontsize=12)
plt.ylabel(r"$\frac{V_{\rm trans}}{c}$", fontsize=12)
plt.axhline(1, color='r', linestyle='--', label=r"$V_{\rm trans} = c$")
plt.title("Variation of transverse velocity with parameters")
plt.legend()
plt.grid()
plt.show()
```

## Question 3:

```python
## Define the function
def gamma(beta):
    return np.sqrt(1 - beta**2)

## Plot the solution
beta = np.arange(0.01, 1, 0.0001) 
plt.figure(figsize = (8, 5), dpi = 150)
plt.plot(beta, 1/gamma(beta))

## Markers and coordinates
special_betas = [0.01, 0.5, 0.9999]
for b in special_betas:
    y_val = 1/gamma(b)
    plt.plot(b, y_val, '*', color='red')
    plt.text(b, y_val, f"({b:.4f}, {y_val:.2f})", fontsize=10, ha='left', va='bottom')

plt.xlabel(r'$\beta$ = $\frac{v}{c}$')
plt.ylabel(r'$\frac{T}{\tau}$')
plt.title("Change in lifetime with the velocity")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
