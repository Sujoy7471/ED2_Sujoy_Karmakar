```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Define force function
def force_fn(t, v0=1):
    e = -1.6e-19
    factor = 9e9 * 2/3
    c = 3e8
    Fx = (factor*e**2/c**3)*np.exp(t/100)*( (10e-4 - 900)*np.sin(30*t) + 3/5 * np.cos(30*t) )
    Fy = 0
    Fz = 0
    return Fx, Fy, Fz

## Define the time arrays
t1 = np.arange(0, 1000, 0.1)
t2 = np.arange(0, 10, 0.01)   # shorter range, higher resolution

## Compute force components
Fx1, Fy1, Fz1 = force_fn(t1)
Fx2, Fy2, Fz2 = force_fn(t2)

## Plot the required quantities
plt.figure(dpi=300, figsize=(12,5))

# Full time range
plt.subplot(1,2,1)
plt.plot(t1, Fx1, label=r"$F_x(t)$", color='tab:blue')
plt.xlabel("Time (t)")
plt.ylabel("Force (N)")
plt.title("Force vs Time (0 to 1000 s)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Short time range
plt.subplot(1,2,2)
plt.plot(t2, Fx2, label=r"$F_x(t)$ (zoomed)", color='tab:orange')
plt.xlabel("Time (t)")
plt.ylabel("Force (N)")
plt.title("Force vs Time (0 to 10 s)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
